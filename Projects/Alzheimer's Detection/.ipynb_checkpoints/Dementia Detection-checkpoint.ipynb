### Import packages

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

### Define constants
We are defining these in order to make the code more readable and modular

```python
IMAGE_SIZE=(176,208)
NUM_CLASSES = 4

BATCH_SIZE = 64
EPOCHS = 60
```

### Import image dataset
Keras API has built-in image loading, refer to API for what each parameter does. We will be using the Image Data Generator Class and the flow from directory method.

```python
train_image_generator = tf.keras.preprocessing.image.ImageDataGenerator(validation_split=0.2)
test_image_generator = tf.keras.preprocessing.image.ImageDataGenerator()
train_generator = train_image_generator.flow_from_directory(
    'Alzheimer_s Dataset/train',
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    subset='training',
    class_mode='categorical')
validation_generator = train_image_generator.flow_from_directory(
    'Alzheimer_s Dataset/train',
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    subset='validation',
    class_mode='categorical')
test_generator = train_image_generator.flow_from_directory(
    'Alzheimer_s Dataset/test',
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical')
```

###### View some sample images

```python
class_names = ['MildDementia', 'ModerateDementia', 'NonDementia', 'VeryMildDementia']
plt.figure(figsize=(10, 10))
images,labels=train_generator.next()
for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].astype("uint8"))
    plt.title(class_names[np.argmax(labels[i])])
    plt.axis("off")
```

### Create model architecture
We will use a convolutional neural network as we are analyzing images, and CNNs work best in that case. We will use batch normalization to make our CNN more efficient and quicker to train, and we have to add dropout to prevent overfitting.

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(*IMAGE_SIZE,3)),
    tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
    tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
    tf.keras.layers.MaxPool2D(),
    tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(32, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(32, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ]),
    tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(64, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(64, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ]),
    tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(128, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(128, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ]),
    tf.keras.layers.Dropout(0.2),
    tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(256, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(256, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ]),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Flatten(),
    tf.keras.Sequential([
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.7)
    ]),
    tf.keras.Sequential([
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5)
    ]),
    tf.keras.Sequential([
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.3)
    ]),
    tf.keras.layers.Dense(NUM_CLASSES, activation='softmax')
    ])
```

```python
model.summary()
```

### Compile model
We will be using an Adam optimizer and Categorical Crossentropy, as our problem is categorical (we have categories as our predicition i.e. Mildly Demented, Not Demented, etc.). We will be using an AUC metric as our dataset is very imbalanced and the normal 'accuracy' will not suffice in our case.

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=[tf.keras.metrics.AUC()])
```

### Train model
Now we will begin training our model, but before we start, we will be using an exponential decay and keras callbacks to make our model converge faster and stop the epochs when the model has achieved optimal accuracy.

```python
def exponential_decay(lr0, s):
    def exponential_decay_fn(epoch):
        return lr0 * 0.1 **(epoch / s)
    return exponential_decay_fn
exponential_decay_fn = exponential_decay(0.01, 20)
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(exponential_decay_fn)
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("alzheimer_model.h5", save_best_only=True)
early_stopping_cb = tf.keras.callbacks.EarlyStopping(patience=15, restore_best_weights=True)
```

```python
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    callbacks=[checkpoint_cb, early_stopping_cb, lr_scheduler],
    epochs=EPOCHS
)
```

### Monitor train and validation accuracy to adjust model

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 3))
ax = ax.ravel()
for i, met in enumerate(['auc', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    if i == 0:
        ax[i].set_ylim(0, 1.1)
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
```

### Evaluate model on test data

```python
model.evaluate(test_generator)
```

##### Our model has an astounding 86% accuracy

### Save model for future use
We will save our model as a json file and our weights as an hierarchical data format

```python
model_json = model.to_json()
with open("Alzheimers.json", "w") as json_file:
    json_file.write(model_json)
model.save_weights("Alzheimers.h5")
```
